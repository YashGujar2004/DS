```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')
df.head()
```

```python
print("\nFeatures and their data types:")
df.info()
```

```python
#use SNS only
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.histplot(df['sepal_length'], kde=True)
plt.title('Distribution of Sepal Length')

plt.subplot(2, 2, 2)
sns.histplot(df['sepal_width'], kde=True)
plt.title('Distribution of Sepal Width')

plt.subplot(2, 2, 3)
sns.histplot(df['petal_length'], kde=True)
plt.title('Distribution of Petal Length')

plt.subplot(2, 2, 4)
sns.histplot(df['petal_width'], kde=True)
plt.title('Distribution of Petal Width')

plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(x='species', data=df)
plt.title('Distribution of Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
# Create boxplots for each numeric feature
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.boxplot(x=df['sepal_length'])
plt.title('Boxplot of Sepal Length')

plt.subplot(2, 2, 2)
sns.boxplot(y=df['sepal_width'])
plt.title('Boxplot of Sepal Width')

plt.subplot(2, 2, 3)
sns.boxplot(y=df['petal_length'])
plt.title('Boxplot of Petal Length')

plt.subplot(2, 2, 4)
sns.boxplot(y=df['petal_width'])
plt.title('Boxplot of Petal Width')

plt.tight_layout()
plt.show()
```

```python
# distributions according to species
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title('Sepal Length by Species')

plt.subplot(2, 2, 2)
sns.boxplot(x='species', y='sepal_width', data=df)
plt.title('Sepal Width by Species')

plt.subplot(2, 2, 3)
sns.boxplot(x='species', y='petal_length', data=df)
plt.title('Petal Length by Species')

plt.subplot(2, 2, 4)
sns.boxplot(x='species', y='petal_width', data=df)
plt.title('Petal Width by Species')

plt.tight_layout()
plt.show()
```
