```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic') 
```

```python
df.describe()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['fare'], bins=30, edgecolor='black') # Adjust 'bins' as needed
plt.title('Distribution of Ticket Price (Fare)')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df.index, df['fare'], color='skyblue', edgecolor='black')
plt.title('Fare Paid by Each Passenger')
plt.xlabel('Passenger Index')
plt.ylabel('Fare')
plt.show()
```
